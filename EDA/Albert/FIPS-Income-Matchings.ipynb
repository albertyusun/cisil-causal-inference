```python
import pandas as pd
```

```python
acs_filepath = '~/Projects/data/CLEAN_King_County_ACS_2019_tract.csv'
acs = pd.read_csv(acs_filepath)
```

```python
acs = acs.set_index('geoid')
```

```python
income_column_names = ['less_than_10000', '10000_to_14999', '15000_to_19999',
       '20000_to_24999', '25000_to_29999', '30000_to_34999',
       '35000_to_39999', '40000_to_44999', '45000_to_49999',
       '50000_to_59999', '60000_to_74999', '75000_to_99999',
       '100000_to_124999', '125000_to_149999', '150000_to_199999',
       '200000_or_more']

```

```python
acs = acs[income_column_names]
```

The poverty level is $12,880. 80\% of the federal poverty level is 10,304. Let's look at the people below yearly 10,000 income. 

```python
acs['Total']= acs[income_column_names].sum(axis=1)
acs['pct_below_80pct_povertyline'] = acs['less_than_10000'] / acs['Total']
acs[['pct_below_80pct_povertyline']].to_csv('geoid_high_poverty.csv')
```

```python
acs_upper_bound = pd.DataFrame(index=acs.index)
for i in range(16):
    acs_upper_bound[acs.columns[i]] = acs.iloc[:,0:i].sum(axis=1)
    
acs_upper_bound['Total'] = acs['Total']

for i in range(16):
    acs_upper_bound.iloc[:,i] -= acs_upper_bound['Total'] / 2
    
acs_upper_bound = acs_upper_bound.T
acs_upper_bound.columns = acs_upper_bound.columns.astype(str)
acs_upper_bound = acs_upper_bound.drop('Total')
acs_upper_bound
```

```python
def find_median(geoid):
    """find median given categories in the indices"""
    
    # return only positive counts for that column
    min_df = acs_upper_bound[acs_upper_bound[geoid] > 0]
    # return the name of the smallest category
    
    print(geoid)
    print(list(min_df[min_df[geoid] == min_df[geoid].min()].index)[0])
    
    return list(min_df[min_df[geoid] == min_df[geoid].min()].index)[0]

all_geoid = list(acs_upper_bound.columns)
all_geoid.remove('53033990100')
all_geoid.remove('53033024200')
all_geoid.remove('53033024602')

# set 53033990100 as NaN because there are no people in that category

for geo_id in all_geoid:
    find_median(geo_id)
```
