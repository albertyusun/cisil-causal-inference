```python
data_path = "/Users/albertsun/Projects/data/LIFT_registry_2022-03-22.csv"
```

```python
import pandas as pd
import numpy as np
```

```python
reg = pd.read_csv(data_path)
```

```python
reg
```

```python
reg['stage'] = reg['card_id'].str[-1:]
reg['id'] = reg['card_id'].str[:-2]
reg
```

```python
reg_grouped_id = reg.groupby(['id']).count()
reg_grouped_id
```

```python
total_reg = reg_grouped_id['stage'].value_counts().plot(kind='bar', title='How many times each person enrolled')
total_reg.set_xlabel("Times enrolled")
total_reg.set_ylabel("Number of People")
```
