# Re-enrollment Counts

```python
reg['stage'] = reg['card_id'].str[-1:]
reg['id'] = reg['card_id'].str[:-2]
reg
```

```python
reg_grouped_id = reg.groupby(['id']).count()[['stage']]
reg_grouped_id
```

```python
total_reg = reg_grouped_id['stage'].value_counts().plot(kind='bar', title='How many times each person enrolled in ORCA-LIFT')
total_reg.set_xlabel("Times enrolled")
total_reg.set_ylabel("Number of People")
total_reg.figure.savefig('total_reg.png', bbox_inches='tight')

```

```python
import matplotlib.pyplot as plt

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
width = .5 # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, label='Men')
ax.bar(labels, women_means, width, bottom=men_means, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
